```python
# 문제3) 각 학교의 총 학생수를 구해 마지막열에 삽입하는 프로그램 작성

import csv
import matplotlib.pyplot as plt
from matplotlib import rc
import numpy as np

f = open('high_school_2019.csv','r',encoding='utf-8')
lines = csv.reader(f)
header = next(lines)

print('-'*100)
print(header[2:])
print('-'*100)


rc('font',family='Malgun Gothic')

list_data = []
for line in lines:
    list_data.append(line[:])

rows = len(list_data)
data = np.zeros((rows,6), dtype='int64')

for i in range(rows):
    for j in range(6):
        data[i][j] = list_data[i][j+2]

data = np.insert(data,6,0,axis=1)

for i in range(rows):
    data[i][6] = data[i][1] + data[i][3] + data[i][5]

print(data)
f.close()
```

```python
# 문제4) 전국 고등학교에 대해 각 학년별 총 학생수를 구하고, 막대그래프로 시각화하는 프로그램을 작성하시오

import csv
import matplotlib.pyplot as plt
from matplotlib import rc
import numpy as np

f = open('high_school_2019.csv','r',encoding='utf-8')
lines = csv.reader(f)
header = next(lines)


rc('font',family='Malgun Gothic')

list_data = []
for line in lines:
    list_data.append(line[:])

rows = len(list_data)
data = np.zeros((rows,6), dtype='int64')

for i in range(rows):
    for j in range(6):
        data[i][j] = list_data[i][j+2]

data = np.insert(data,6,0,axis=1)

for i in range(rows):
    data[i][6] = data[i][1] + data[i][3] + data[i][5]


xdata = ['1학년','2학년','3학년']
total_student = [0,0,0]

for i in range(rows):
    total_student[0] += data[i][1]
    total_student[1] += data[i][3]
    total_student[2] += data[i][5]

print("\t\t1학년 총 학생수 : ",total_student[0])
print("\t\t2학년 총 학생수 : ",total_student[1])
print("\t\t3학년 총 학생수 : ",total_student[2])
print()
plt.bar(xdata,total_student)
plt.title("전국 고등학교 학년별 총 학생수")
plt.show()




f.close()
```
