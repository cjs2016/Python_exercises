```python
import csv

file1 = open('month_temp.csv','r',encoding='utf-8')
file2 = open('month_temp2.csv','w',encoding='utf-8', newline='')  #newline='' os에 따른 개행문자 변환을 하지 말라는 뜻

lines = csv.reader(file1)
wr = csv.writer(file2)  

wr.writerow(['일자','최저기온(°C)','최고기온(°C)','일교차(°C)'])
next(lines)

for line in lines:
    diff = float(line[4]) - float(line[3])
    diff = format(diff,'.1f')
    wr.writerow([line[1],line[3], line[4],diff])

print('파일 쓰기 완료')
file1.close()
file2.close()

```

```python
import csv

f = open('month_temp.csv','r',encoding='utf-8')
lines = csv.reader(f)
next(lines)
sum = 0
for line in lines:
    if int(line[1][-2:]) <= 10:
        sum += float(line[2])
avg = sum / 10
print('10일간 평균 기온 : %.1f' % avg)
f.close()
```

```python
import csv

f = open('jeju_2019.csv','r',encoding='utf-8')
lines = csv.reader(f)

header = next(lines)
print(header[1],header[2],header[3])

month = 1
sum = 0
num_day = 0

for line in lines:
    if line[1] == '서귀포' and int(line[2][5:7]) == month:
        sum += float(line[3])
        num_day += 1
avg = sum/num_day
print("%d월 일수: %d"%(month,num_day))
print("%d월 최저기온 평균 : %.1f"%(month,avg))
    

f.close()
```

```python
import csv

f = open('pharm_2019.csv','r',encoding='utf-8')
lines = csv.reader(f)

header = next(lines)
print('번호',header[0],header[2],sep=',')
data = []

for line in lines:
    if line[1] == '용인수지구':
        tmp = '%s/%s'%(line[0],line[2])
        data.append(tmp)

data.sort()
number = 1
for x in data:
    print('%d. %s'%(number,x))
    number += 1

f.close()    
```
