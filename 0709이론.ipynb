```python
import matplotlib.pyplot as plt
from matplotlib import rc

rc('font',family='Malgun Gothic')
font1 = {'size':18, 'color':'green'}
font2 = {'size':12}
plt.plot(['kim','lee','kang'],[85,88,90])
plt.title('English Score of three students', fontdict=font1)
plt.xlabel('Student Name',fontdict=font2)
plt.ylabel('Score',fontdict=font2)
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib import rc

rc('font',family='Malgun Gothic')
font1 = {'size':18, 'color':'green'}

xdata = ['','','']
ydata1 = [90,85,88]
ydata2 = [83,88,91]

plt.plot(xdata,ydata1,label='',color='magenta',linestyle='-', marker='s') # red, green, blue,cyan,magenta,yello,black,white
plt.plot(xdata,ydata2,label='', color='#444444', linestyle=':',marker='d') # - -- -., :
                                                                               # marker o, x,s,d
plt.legend(loc='upper center')  #upper center, lower center, best, center left,center,

plt.title('  ,  ')
plt.show()
```

```python
import matplotlib.pyplot as plt

xdata = list(range(30))
ydata = [ 2 * x for x in range(30)]

plt.plot(xdata,ydata)
plt.xlim(0,35)
plt.ylim(0,70)
plt.xticks(list(range(0,35,2)))
plt.yticks(list(range(0,70,5)))
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib import rc

rc('font',family='Malgun Gothic')
pets = ['','','','']
portion = [18,3.4,3.1,75.5]


#portion, explode  0.1= 10%, labels, autopct  . %   '%1.1f%%'
#shadow , startangle x  

plt.pie(portion,explode=(0,0.1,0,0),labels = pets,autopct='%1.1f%%',shadow=False, startangle=0)
plt.show()
```
