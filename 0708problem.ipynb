```python
import csv

f = open('hospital_2019.csv','r',encoding='utf-8')
lines = csv.reader(f)
header = next(lines)

# 문제 2. 제주도에서 2016~2018년 3년 간 개원한 의원 수를 구하는 프로그램을 작성하시오

sum = 0
for line in lines:
    if line[2] == '제주':
        if not line[5] :
            line[5] = 0
        if (int(line[5]) >= 20160101) and (int(line[5]) <= 20181231):
            sum += int(line[6])

print('2016~2018년 3개년간 개원한 제주도의 의원 수 : %d개'%sum)

f.close()     
```

```python
import csv

f = open('hospital_2019.csv','r',encoding='utf-8')
lines = csv.reader(f)
header = next(lines)

# 문제 4. 강원 지역에서 2019년 개원한 한의원에 대해 한의원명, 주소, 총 의사수의 목록을 출력하시오
"""
line[0] 병원명
line[1] 종류
line[2] 지역
line[3] 주소
line[4] 전화번호
line[5] 개설일자
line[6] 총의사수
"""
print('번호','한의원명','전화번호','총의사수')

num = 1
for line in lines:
    if (line[2] == '강원') and ('한의원' in line[0]):
        if (int(line[5]) >= 20190101) and (int(line[5]) <= 20191231): 
            print(f'{num}.{line[0]}/{line[4]}/{line[6]}')
            num = num + 1
            
f.close()     
```

```python
import csv

f = open('hospital_2019.csv','r',encoding='utf-8')
lines = csv.reader(f)
header = next(lines)

# 문제 4. 부산의 병원명에 '삼성'이 들어가는 병원에 대해 총 병원수와 병원명, 종류, 주소, 개설일자의 목록을 출력하시오
"""
line[0] 병원명
line[1] 종류
line[2] 지역
line[3] 주소
line[4] 전화번호
line[5] 개설일자
line[6] 총의사수
"""
print('번호','병원명','종류','주소','개설일자')

num = 1
for line in lines:
    if (line[2]=='부산')and('삼성' in line[0]) :
        print(f'{num}.{line[0]}/{line[1]}/{line[3]}/{line[5]}')
        num = num + 1
            
f.close()     
```
