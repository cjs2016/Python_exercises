```python
import matplotlib.pyplot as plt
from matplotlib import rc
import csv

f = open('cafe_2year.csv','r',encoding='utf-8')
lines = csv.reader(f)
next(lines)

base_month = ['03','06','09','12']

xdata = []
ydata = []

for line in lines:
    if line[0][-2:] in base_month:
        xdata.append(line[0])
        ydata.append(line[1])

plt.bar(xdata,ydata)


rc('font',family='Malgun Gothic')
font1 = {'size':18, 'color':'green'}

plt.title('전국의 월별 카페수',fontdict=font1)
plt.xlabel('년월')
plt.ylabel('카페수')

plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib import rc
import csv

f = open('doctor_2019.csv','r',encoding='utf-8')
lines = csv.reader(f)
next(lines)

area = ['서울','부산','대구','인천','대전','광주','울산']

rows, cols = len(area), 5
array = [[0 for _ in range(cols)] for _ in range(rows)]

for i in range(len(array)):
    array[i][0] = area[i]
                
for line in lines:
    city = line[0]
    if city not in area:
        continue
    else:
        loc = area.index(city)
    array[loc][1] += int(line[2])
    array[loc][2] += int(line[3])
    array[loc][3] += int(line[4])
    array[loc][4] += int(line[5])


print('지역','일반의','인턴','레지턴트','전문의')
for i in range(len(array)):
    for j in range(len(array[i])):
        print("%-5s"%array[i][j],end=' ')
    print()    
            
f.close()    
                
```

```python
import csv

file1 = open('doctor2.csv','w',encoding='utf-8', newline='') 
wr = csv.writer(file1)  

wr.writerow(['지역','일반의','인턴','레지턴트','전문의'])
wr.writerows(array)

print("파일쓰기완료")
file1.close()
```

```python
import matplotlib.pyplot as plt
from matplotlib import rc
import csv

f = open('doctor2.csv','r',encoding='utf-8')
lines = csv.reader(f)
next(lines)

area_data = []
doc1_data = [] #일반의
doc2_data = [] #인턴
doc3_data = [] #레지던트
doc4_data = [] #전문의


for line in lines:
    area_data.append(line[0])
    doc1_data.append(int(line[1])) #일반의
    doc2_data.append(int(line[2])) #인턴
    doc3_data.append(int(line[3])) #레지던트
    doc4_data.append(int(line[4])) #전문의


plt.plot(area_data,doc1_data,label='일반의',color='magenta',linestyle='-', marker='o') 
plt.plot(area_data,doc2_data,label='인턴', color='blue', linestyle='--',marker='x')
plt.plot(area_data,doc3_data,label='레지던트', color='yellow', linestyle=':',marker='s')
plt.plot(area_data,doc4_data,label='전문의', color='purple', linestyle='-.',marker='d')

plt.legend(loc='upper right')

rc('font',family='Malgun Gothic')
font1 = {'size':18, 'color':'green'}

plt.title('서울 및 광역시 종합병원 전문의 수',fontdict=font1)
plt.xlabel('시')
plt.ylabel('의사수')

plt.show()
f.close()
```

```python
import matplotlib.pyplot as plt
from matplotlib import rc
import csv

f = open('doctor2.csv','r',encoding='utf-8')
lines = csv.reader(f)
next(lines)

data = next(lines)
print(data)

portion = [] 
for i in range(1,5):
    portion.append(int(data[i]))

doc_type = ['일반의','인턴','레지던트','전문의']
plt.pie(portion,explode=(0,0.1,0,0),labels = doc_type ,autopct='%1.1f%%',shadow=False, startangle=90)
plt.show()


f.close()
```
