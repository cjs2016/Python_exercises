```python
#실습) 예제-14에 최소 학급수/학생수/교사수를 가지는 서울 소재 학교의 이름과 인원수를 출력
# 서울 소재 초등학교 만 발췌해서 Ndarray 생성 먼저 할 것

import csv
import numpy as np

f = open('school_2019.csv','r',encoding='utf-8')
lines = csv.reader(f)

header = next(lines)

list_data = []
for line in lines:
    if '서울' in line[0]:
        list_data.append(line[:])

length = len(list_data)
data = np.zeros((length,3), dtype='int64')
for i in range(length):
    for j in range(3):
        data[i][j] = int(list_data[i][j+2])

min_index = np.argmin(data,axis = 0)
print(min_index)
min_class = list_data[min_index[0]][1] 
min_num = list_data[min_index[0]][2]
print('최소 학급수를 가지는 학교명: {0}, 학급수: {1}'.format(min_class,min_num))

min_student = list_data[min_index[1]][1] 
min_num = list_data[min_index[1]][3]
print('최소 학생수를 가지는 학교명: {0}, 학생수: {1}'.format(min_student,min_num))

min_teacher = list_data[min_index[2]][1] 
min_num = list_data[min_index[2]][4]
print('최소 교사수를 가지는 학교명: {0}, 교사수: {1}'.format(min_teacher,min_num))


```

```python
import csv
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
rc('font',family='Malgun Gothic')

f = open('school_2019.csv','r',encoding='utf-8')
lines = csv.reader(f)
header = next(lines)
print(header)

schools = {}
for line in lines:
    if('광주' in line[0]) and ('마재' in line[1]):
        schools.update({'마재':[line[2],line[3],line[4]]})
    if('울산' in line[0]) and ('약사' in line[1]):
        schools.update({'약사':[line[2],line[3],line[4]]})
    if('용인' in line[0]) and ('정평' in line[1]):
        schools.update({'정평':[line[2],line[3],line[4]]}) 
    if('제주' in line[0]) and ('도평' in line[1]):
        schools.update({'도평':[line[2],line[3],line[4]]})     

print(schools)

school = list(schools.values())

data = np.zeros((4,3),dtype='int64')

for i in range(len(schools)):
    for j in range(3):
        data[i][j] = school[i][j]

print(data)

data = np.insert(data,2,0,axis=1)  #학급당 학생수
data = np.insert(data,4,0,axis=1)  #교사 1인당 학생수
print(data)
print(data.shape[0])  #0행의 수, 1 열의수
row = data.shape[0]

"""
0 학급수
1 학생수
2 학급당 학생수
3 교사수
4 교사1인당 학생수

"""

for i in range(row):
    data[i][2] = round(data[i][1]/data[i][0])   
    data[i][4] = round(data[i][1]/data[i][3])

print(data)

xdata = ['마재','약사','정평','도평']
plt.plot(xdata,data[:,1],label='학생수',color='red', linestyle='--',marker='x')
plt.plot(xdata,data[:,3],label='교사수',color='blue', linestyle=':',marker='d')
plt.title('학생수와 교사수')
plt.legend(loc='best')
plt.show()

```
