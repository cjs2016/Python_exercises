```python
import csv

f = open('school_2019.csv','r',encoding='utf-8')
lines = csv.reader(f)

header = next(lines)
print(header)

list_data = []
for line in lines:
    list_data.append(line[:])

print('-'*50)
length = len(list_data)

data = np.zeros((length,3), dtype='int64')

for i in range(length):
    for j in range(3):
        data[i][j] = int(list_data[i][j+2])


print(data)
print('-'*50)

max_index = np.argmax(data,axis=0)
print(max_index)
max_class = list_data[max_index[0]][1]
print(max_class)
f.close()
```

```python
#실습) 예제-14에 최소 학급수/학생수/교사수를 가지는 서울 소재 학교의 이름과 인원수를 출력
# 서울 소재 초등학교 만 발췌해서 Ndarray 생성 먼저 할 것

import csv
import numpy as np

f = open('school_2019.csv','r',encoding='utf-8')
lines = csv.reader(f)

header = next(lines)

list_data = []
for line in lines:
    if '서울' in line[0]:
        list_data.append(line[:])

length = len(list_data)
data = np.zeros((length,3), dtype='int64')
for i in range(length):
    for j in range(3):
        data[i][j] = int(list_data[i][j+2])

min_index = np.argmin(data,axis = 0)
print(min_index)
min_class = list_data[min_index[0]][1] 
min_num = list_data[min_index[0]][2]
print('최소 학급수를 가지는 학교명: {0}, 학급수: {1}'.format(min_class,min_num))

min_student = list_data[min_index[1]][1] 
min_num = list_data[min_index[1]][3]
print('최소 학생수를 가지는 학교명: {0}, 학생수: {1}'.format(min_student,min_num))

min_teacher = list_data[min_index[2]][1] 
min_num = list_data[min_index[2]][4]
print('최소 교사수를 가지는 학교명: {0}, 교사수: {1}'.format(min_teacher,min_num))


```
